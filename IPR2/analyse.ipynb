```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('dermatology_database_1.csv')
```

```python
data.columns
```

```python
ax = plt.subplots(figsize=(10,10)) 
ax = sns.heatmap(round(abs(data.corr()), 1), annot=True)
```

```python
data = data.drop(['polygonal_papules', 'oral_mucosal_involvement', 'oral_mucosal_involvement', 'melanin_incontinence', 'focal_hypergranulosis',
                 'vacuolisation_damage_basal_layer', 'saw_tooth_appearance_retes', 'follicular_horn_plug', 'perifollicular_parakeratosis',
                 'band_like_infiltrate', 'clubbing_rete_ridges', 'thinning_suprapapillary_epidermis', 'scalp_involvement', 'exocytosis', 'elongation_rete_ridges'], axis=1)
```

```python
X, y = train_test_split(data, test_size=0.2, random_state=0)
```

```python
X.to_excel('train.xlsx', engine='xlsxwriter', sheet_name='train')
y.to_excel('test.xlsx', engine='xlsxwriter', sheet_name='test')
```
